```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_path = r'D:\codelearning\dataanalysis\NYC_ab\AB_NYC_2019.csv'
raw_data = pd.read_csv(data_path)
select_cols = ['neighbourhood_group', 'neighbourhood', 'latitude', 'longitude', 'room_type', 'price', 'minimum_nights']
data = raw_data[select_cols]
```

```python
# 按照 "neighbourhood_group" 列分组，计算平均价格
grouped_df = data.groupby('neighbourhood_group')['price'].mean()

# 设置颜色和透明度
colors = ['blue', 'green', 'red']
alpha = 0.5

# 绘制彩色柱状图
plt.bar(grouped_df.index, grouped_df.values, color=colors, alpha=alpha)
# 添加标签
for i, v in enumerate(grouped_df.values):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom')

# 设置图形标题和轴标签
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')

# 显示图形
plt.show()
```
